```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
people = pd.read_csv('people.csv')
```

```python
people.head()
```

```python
x = people['Gpa']
y = people['#Friends']

y_points = np.array([y.loc[y.idxmin()], y.loc[y.idxmax()]])
x_points = np.array([x.loc[x.idxmin()], x.loc[x.idxmax()]])

plt.scatter(x,y)
plt.plot(x_points, y_points, c='red')
plt.grid()
plt.xlabel('Gpa')
plt.ylabel('# of Friencs')
plt.title('Gpa v.s. # of Friends')
plt.show()
```

```python
corr = people.corr(numeric_only=True)

corr
```
