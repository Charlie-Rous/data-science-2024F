# Cleaning data

Items Covered:

```python
# Vizualize null data
null_mask_rows = grads.isnull().any(axis=1)

null_rows = grads[null_mask_rows]

null_rows.head()
```

isnull created a boolean flag for all rows that had null values. By passing this boolean flag into the dataframe I found that only the Food Science major had missing values. 

```python
# Replace null data

categorical_cols = ['Major', 'Major_category']

grads_no_missing = grads.fillna(grads.median(skipna=True, numeric_only=True))

grads_no_missing.loc[grads_no_missing['Major'] == 'FOOD SCIENCE']
```

```python
grads_no_missing.rename(columns = {'ShareWomen': 'Percentage_Women'})
```

Using fillna, the missing values in food science were replaced with the median value from that column.

```python
Stem_majors = ['Engineering','Physical Sciences', 'Health', 
               'Computer & Mathematics', 'Biology & Life Science', 
               'Agriculture & Natural Resources']

Non_stem_majors = ['Buisiness', 'Law & Public Policy', 'Arts', 
                   'Social Science', 'Education', 'Humanities & Liberal Arts',
                      'Psychology & Social Work', 'Industrial Arts & Consumer Services',
                         'Communications & Journalism', 'Interdisciplinary']

grads_no_missing['STEM'] = ['STEM' if i in Stem_majors else 'Non-STEM' for i in grads_no_missing['Major_category'] ]

grads_no_missing['STEM'].value_counts()
```

I created a new column in the dataframe called STEM which defined whether or not the major was a STEM major. From Value_Counts I can see that there are 98 non-stem majors and 75 stem majors.

# Salary Analysis
Items Coverd:
* groupby()
* .agg()
* .apply()
* lambda
* loc()
* idxmax()
* mean, median, min, max,, var, std


```python
grads_no_missing.sort_values('Median', ascending=False)[:10]
```

```python
grads_no_missing.sort_values('Median', ascending=True)[:10]
```

```python
# Max median salary by major in both stem and non-stem majors

highest_salary = grads_no_missing.groupby(['STEM']).apply(lambda df: df.loc[df.Median.idxmax()])

highest_salary
```

This cell shows that the highest median salary in non-stem majors is $62,000 (Actuarial Science) and the highest median salary in stem majors is $110,000 (Petroleum Engineering)

```python
salaries = grads_no_missing.groupby('STEM').Median.agg(Average='mean', STD='std', Minimum = 'min', Med = 'median', Maximum='max', Variance='var')

salaries
```

This cell creates columns which calculate the mean, standard deviation, minimum, median, variance, and maximum for both STEM and non-STEM majors. The minimum and median slaries are fairly similar for both stem and non-stem majors however the maximum median salaries for stem majors is much higher than the maximum salries for non-stem majors. The stem-majors also have more variation in their salaries.

```python
grads_no_missing['Income_Range'] = pd.qcut(grads_no_missing['Median'], 3, labels=['low', 'med','high'])

grads_no_missing.groupby('STEM').Income_Range.value_counts()
```


This cell uses qcut, which divides a column up into even bins and allows you to label these bins. In this case I divided the median income column into 3 bins and labled them low, med, and high. I then sorted this values by stem and non-stem majors and looked at the value counts. Stem majors had the vast majority of high incomes and a much lower proportion of low incomes. Non-stem majors were the opposite.

Its clear to see from this analysis that typically STEM majors have better median salaries than non-stem majors.

# Gender Analysis


```python
grads_no_missing = grads_no_missing.assign(Percentage_Men= lambda df: df.Men / df.Total)

grads_no_missing.groupby('STEM').Percentage_Men.agg('mean')
```

This cell assigned a new column to calculate the percentage of men in each major. I then used groupby to find the average percentage of men in stem majors and non-stem majors. There is a clear gender gap between stem and non-stem majors. Stem majors have 54% men whereas non-stem majors have only 42% men. 

```python
lowest_men = grads_no_missing.groupby(['STEM']).apply(lambda df: df.loc[df.Percentage_Men.idxmin()])

lowest_men
```

The most female dominated majors in STEM and non-stem are Communication Disorders Sciences and Services, and Early Childhood Education respectivly.

```python
highest_men = grads_no_missing.groupby(['STEM']).apply(lambda df: df.loc[df.Percentage_Men.idxmax()])

highest_men
```

The most Male dominated fields are Military technologies and Mechanical Engineering Related Technologies.
